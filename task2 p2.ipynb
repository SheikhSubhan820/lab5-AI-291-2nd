```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns

# Define the data directly as a dictionary
data = {
    'area': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500],  # Area in sq ft
    'price': [150000, 180000, 210000, 240000, 270000, 300000, 330000, 360000]  # Price in $
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Features (X) and Target (y)
X = df[['area']]  # Independent variable (area)
y = df['price']   # Dependent variable (price)

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict the price for specific areas
areas_to_predict = [5000, 8000, 9000]
predicted_prices = model.predict([[area] for area in areas_to_predict])

# Display the predictions
print("Predicted Prices for the given areas:")
for area, price in zip(areas_to_predict, predicted_prices):
    print(f"Area: {area} sq ft, Predicted Price: ${price:.2f}")

# Plotting the original data and the regression line
plt.figure(figsize=(10, 6))

# Scatter plot of the original data points
sns.scatterplot(x=df['area'], y=df['price'], color='blue', label='Actual Prices')

# Plot the regression line
plt.plot(df['area'], model.predict(X), color='red', linewidth=2, label='Regression Line')

# Plot the predicted points for specific areas
plt.scatter(areas_to_predict, predicted_prices, color='green', marker='o', s=100, label='Predicted Points')

# Add labels and title
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Area vs Price')
plt.legend()

# Show the plot
plt.show()
```
